```python
# Question 1. import numpy, pandas, and stats models
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
# Question  2. Load the data into a pandas DataFrame. Calculate summary statistics and make sure everything looks sensible.
df = pd.read_csv("wellness_data.csv")
df.describe()
```

Question 2 comments:

```python
# Question 3. Measure associaiton between X and Y variable
# a. Calculate average of spend_year1 among people with wellness==1 and wellness==0

# Average of spend_year1 when wellness == 1
avg_wellness_1 = df[df['wellness'] == 1]['spend_year1'].mean()
print("Year 1 spending average when wellness = 1: ", avg_wellness_1)

# Average of spend_year1 when wellness == 0
avg_wellness_0 = df[df['wellness'] == 0]['spend_year1'].mean()
print("Year 1 spending average when wellness = 0: ", avg_wellness_0)

# b. Estimate a regression of spend_year1 on wellness.


# Estimate a regression of spend_year1 on wellness.
y = df['spend_year1']
X = df['wellness']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Question 7. Investgiate whether diabetes_pre is a confound, and adjust for it
# a. Find the fraction of people with diabetes_pre  who use the wellness program,
#     and the fraction of people without diabtes_pre  who use the wellness program.

frac_with_diabetes = df[df['diabetes_pre'] == 100]['wellness'].mean()
frac_without_diabetes = df[df['diabetes_pre'] == 0]['wellness'].mean()
print("Fraction of people with diabetes_pre who use the wellness program: ", frac_with_diabetes)
print("Fraction of people without diabetes_pre who use the wellness program: ", frac_without_diabetes)

# b. Find average spending among people with diabetes_pre, and among people without.

avg_spending_with_diabetes = df[df['diabetes_pre'] == 100]['spend_year1'].mean()
avg_spending_without_diabetes = df[df['diabetes_pre'] == 0]['spend_year1'].mean()
print("Average spending among people with diabetes_pre: ", avg_spending_with_diabetes)
print("Average spending among people without diabetes_pre: ", avg_spending_without_diabetes)

# C. Measure the association between wellness and spend_year1 among people with
#   diabetes_pre (as in step 3) and, separately, without diabetes_pre.


# d. To adjust for diabetes_pre, estimate a regression of spend_year1 on
#    wellness and diabetes_pre. What is the coefficient on wellness now?

y = df['spend_year1']
X = df[['diabetes_pre', 'wellness']]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

Question 7 comments/findings:

- Fraction of people with diabetes_pre who use the wellness program:  0.49

The average spending for people with diabetes is considerabily higher than the average spending for those without. Also, People with diabetes are less likely to opt into the wellness program.

```python
# Question 8. Adjust for all potential confounds in the data
# Specifically, estimate a regression of spend_year1 on wellness,
# and also spend_pre, the age variables, and diabetes_pre.
# What is the coefficient on wellness?

y = df['spend_year1']
X = df[['wellness', 'age50', 'age37_49', 'age1836', 'diabetes_pre', 'spend_pre']]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())

```

Question 8 comments:

 - The coefficient on wellness is 13.585. Considering all other variables as contant, people who join the wellness program are associated with an increase of 13.58 dollars in year 1 spending.
