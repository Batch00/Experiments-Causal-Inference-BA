```python
# Carson Batchelor
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dansacks/gb740/main/exam_data_2024_1111.csv")
df.head()
```

```python
table = df.groupby("arm")["revenue"].agg(mean="mean", sem = "sem")

table["ate"] = table["mean"] - table["mean"]["baseline"]
table["lift"] = table["ate"] / table["mean"]["baseline"]

table["se_ate"] = np.sqrt(table["sem"]**2 + table["sem"]["baseline"]**2)
table["lb"] = table["ate"] - 1.96 * table["se_ate"]
table["ub"] = table["ate"] + 1.96 * table["se_ate"]


table = table.drop(columns = ["sem", "se_ate"])
table
```
